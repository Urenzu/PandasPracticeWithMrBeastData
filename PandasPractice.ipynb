# Subsetting Data

```python
df.columns # Listing out every column we have within the DataFrame
```

```python
# We can create a new DataFrame with only particular columns that we want. (Subsetting Columns)
df2 = df[['title', 'description', 'publishTime', 'duration_seconds', 'viewCount', 'likeCount', 'commentCount']]
```

```python
df2.head()
```

```python
df2.shape # Seeing rows and then columns
```

```python
# (Subsetting Rows) Subsetting using loc
df_subset1 = df2.loc[df2["viewCount"] > 1_000_000] # Showing only rows where the video view count is above one million.
```

```python
# Subsetting Rows but using query method.
df_subset2 = df2.query("viewCount > 1000000")
```

```python
df_subset2
```

```python
# Gets rid of all rows where the viewCount column had a NaN.
df2 = df2.loc[~df2["viewCount"].isna()]
df2 = df2.loc[~df2["likeCount"].isna()]
```

# Casting dtypes

(Changing a data type from it's original type)

```python
df2.dtypes
```

```python
# When the columns don't have any null values you can convert them to another type. In this instance we are converting floats to ints.
df2["viewCount"] = df2["viewCount"].astype("int")
df2["likeCount"] = df2["likeCount"].astype("int")
```

```python
df2.dtypes
```

```python
# We just converted an object or string type to a useable datetime type using pandas.
df2["publishTime"] = pd.to_datetime(df2["publishTime"])
```

```python
df2["likeCount"] = pd.to_numeric(df2["likeCount"].astype("str"))
```

# Creating a new column

```python
# Creating a new column on the DataFrame that is like to view ratio.
df2["like_to_view_ratio"] = df2["likeCount"] / df2["viewCount"]
```

```python
df2.head()
```

# Creating a new row

```python
df_to_append = df2.tail(1)
```

```python
df_concat = pd.concat([df2, df_to_append]) # we simply duplicated a row on the data, but this is an example to show how you can add a row to the dataFrame.
df_concat
```
